### Zero-shot Text Classification

```python
from transformers import pipeline

# Load the zero-shot pipeline once
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def zero_shot_classify(text, labels, multi_label=False):
    """
    Performs zero-shot text classification.

    Args:
        text (str): The input text to classify.
        labels (list): List of candidate labels.
        multi_label (bool): If True, allows multiple labels per text.

    Returns:
        dict: Classification results with labels and scores.
    """
    result = classifier(text, candidate_labels=labels, multi_label=multi_label)

    print(f"\nText: {text}")
    print("Classification Results:")
    for label, score in zip(result['labels'], result['scores']):
        print(f"  {label:<15} → {score:.4f}")

    return result
```

```python
# ---------- Example Usage ----------

# Example 1: Single-label classification
zero_shot_classify(
    "I love watching cricket with my friends on weekends.",
    ["sports", "politics", "technology", "food"]
)

# Example 2: Multi-label classification
zero_shot_classify(
    "The iPhone 15 has an amazing camera but is too expensive.",
    ["technology", "sports", "business", "entertainment"],
    multi_label=True
)
```

```python
zero_shot_classify(
    "She sells seashells by the seashore, the shells she sells are surely seashells.",
    ["tongue twister", "story", "quote", "random text"]
)
```
