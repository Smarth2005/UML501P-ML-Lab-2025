### Image Captioning using Hugging Face Transformer (Computer Vision + NLP)

```python
from transformers import pipeline
from PIL import Image
import requests
from IPython.display import display, HTML

# Load pipeline once
captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
def generate_caption(image_path=None, url=None, show=True):
    """
    Generates a caption for the given image and optionally displays it.
    """
    # Load image
    if image_path:
        image = Image.open(image_path)
    elif url:
        image = Image.open(requests.get(url, stream=True).raw)
    else:
        raise ValueError("Please provide either image_path or url")

    # Generate caption
    result = captioner(image)
    caption = result[0]['generated_text']

    # Display image + caption inline (Jupyter/Colab)
    if show:
        display(HTML(
        f"<img src='{url}' width='400'>"
        f"<p style='font-size:16px;'><b>Generated Caption:</b> {caption}</p>"
        ))

    # return caption
```
